# Task 4: Image To Pencil Sketch With Python
## Description Of The Task :
We need to read the image in RBG format and then convert it into the greyscale image. This will be turn an image into 
 a classic black and white photo. Then the next thing to do is invert the greyscale image also called negative image,
this will be our greyscale inverted image. Inversion can be used to enhance details. Then we can finally create the pencil 
 sktech by mixing the grayscale image with the inverted blurry image. This can be done by dividing the grayscale image by the inverted blurry image. Since images are just , we can easily do this programmatically using the divide function from the cv2 library in Python.

# Importing Libraries

```python
import numpy as np
import cv2
from PIL import Image
from IPython.display import display
```

# Read The Image In RBG Format

```python
img = cv2.imread("Shivaji.jpg")
img = cv2.resize(img,(250,300))
RBG_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
display(Image.fromarray(RBG_img))
```

# RBG Image Into Greyscale Image (White and Black Photo)

```python
grey_img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
display(Image.fromarray(grey_img))
```

# Inverted Greyscale Image (Negative Image)

```python
inverted_grey_img = 255-grey_img
display(Image.fromarray(inverted_grey_img))
```

# Convert Into Negative Image

```python
Negative_img = cv2.bitwise_not(grey_img)
display(Image.fromarray(Negative_img))
```

# Smoothen The Image 

```python
Smooth_img = cv2.GaussianBlur(Negative_img,(5,5),0,0)
display(Image.fromarray(Smooth_img))
```

# Convert Into Pencil Sketch

```python
invert_blur = 255 - Smooth_img
Pencil_Sketch = cv2.divide(grey_img,invert_blur,scale=250.0)
display(Image.fromarray(Pencil_Sketch))
```

# Combination Of Original And Pencil Sketch Image

```python
display(Image.fromarray(RBG_img))
display(Image.fromarray(Pencil_Sketch))
```
